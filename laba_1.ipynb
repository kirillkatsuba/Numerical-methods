# Задача 1
## $\sum\limits_{n=0}^\infty\frac{36}{n^2+7n+10}$

```python
import matplotlib.pyplot as plt
import numpy as np
real_sum = 13

# Вычисляем частичную сумму ряда
def series_sum(n: int):
    res = 0
    for i in range(n+1):
        res += 36/(i**2 + 7*i + 10)
    return res

# Считаем количество верных значащих цифр
def significant_digit(a: float):
    error = abs(13 - a)
    sig_digits = 0
    t = 1
    while error <= (10 ** t):
        t -= 1
        sig_digits += 1
    return sig_digits

# Кладем частичные суммы, погрешности и верные цифры для каждого n в массив
part_sum = []
abs_error = []
signif_digits = []
for i in range(1,6):
    part_sum.append(series_sum(10**i))
    abs_error.append(abs(13 - part_sum[-1]))
    signif_digits.append(significant_digit(series_sum(10**i)))

# Выводим ответы
for i in range(5):
    print(f'Частичная сумма, n = {10**(i+1)}: {part_sum[i]}')
print('\n')
for i in range(5):
    print(f'Погрешность, n = {10**(i+1)}: {abs_error[i]}')
print('\n')
for i in range(5):
    print(f'Верные значащие цифры, n = {10**(i+1)}: {signif_digits[i]}')
```

```python
# Гистограмма верных значащих цифр
plt.bar(['10', '100', '1000', '10000', '100000'], signif_digits)
plt.xlabel('Количество слагаемых в сумме ряда')
plt.ylabel('Количество верных значащих цифр')
plt.title('Гистограмма верных значащих цифр')
plt.savefig('/Users/kirill/edu/3course/Numerical_methods/laba_1/hist_dig.png')
```

```python
# Гистограмма абсолютной погрешности
plt.bar(['10', '100', '1000', '10000', '100000'], abs_error)
plt.xlabel('Количество слагаемых в сумме ряда')
plt.ylabel('Абсолютная погрешность')
plt.title('Гистограмма абсолютной погрешности')
plt.savefig('/Users/kirill/edu/3course/Numerical_methods/laba_1/hist_error.png')
```

# Задача 2
$$ 
\begin{equation*}
A =
\begin{pmatrix}
9 & 5 & 6\\
17 & 9 & 11\\
7 & 4 & 5
\end{pmatrix}
\end{equation*}
$$

```python
# Считаем определитель
def determinat(A: np.ndarray):
    res = A[0,0]*A[1,1]*A[2,2] + A[0,1]*A[1,2]*A[2,0] + A[0,2]*A[1,0]*A[2,1]
    res -= A[0,2]*A[1,1]*A[2,0] + A[0,0]*A[1,2]*A[2,1] + A[0,1]*A[1,0]*A[2,2]
    return res

# Считаем количество верных значащих цифр
def significant_digit(det: float, det_real: float):
    error = abs(det_real - det)
    sig_digits = 0
    t = 0
    while error <= (10 ** t):
        t -= 1
        sig_digits += 1
    return sig_digits
    
# Инициализируем матрицу
A = np.array([[9., 5., 6.],[17., 9., 11.], [7., 4., 5.]])
Det_real = determinat(A)
print(f'Определитель исходной матрицы равен {Det_real}\n')

# Меняем только первый диагональный элемент
A = np.array([[9., 5., 6.],[17., 9., 11.], [7., 4., 5.]])
A[0][0] *= 1.01
Det = determinat(A)
print(f'Определитель матрицы с погрешностью у первого диагонального элемента {Det}')
print(f'Количество верных значащих цифр: {significant_digit(Det, Det_real)}')
print(f'Относительная погрешность: {abs(Det_real - Det) / abs(Det_real)}\n')

# Меняем только второй диагональный элемент
A = np.array([[9., 5., 6.],[17., 9., 11.], [7., 4., 5.]])
A[1][1] *= 1.01
Det = determinat(A)
print(f'Определитель матрицы с погрешностью у второго диагонального элемента {Det}')
print(f'Количество верных значащих цифр: {significant_digit(Det, Det_real)}')
print(f'Относительная погрешность: {abs(Det_real - Det) / abs(Det_real)}\n')

# Меняем только третий диагональный элемент
A = np.array([[9., 5., 6.],[17., 9., 11.], [7., 4., 5.]])
A[2][2] *= 1.01
Det = determinat(A)
print(f'Определитель матрицы с погрешностью у третьего диагонального элемента {Det}')
print(f'Количество верных значащих цифр: {significant_digit(Det, Det_real)}')
print(f'Относительная погрешность: {abs(Det_real - Det) / abs(Det_real)}\n')

# Меняем первый и второй диагональные элементы
A = np.array([[9., 5., 6.],[17., 9., 11.], [7., 4., 5.]])
A[0][0] *= 1.01
A[1][1] *= 1.01
Det = determinat(A)
print(f'Определитель матрицы с погрешностью у первого и второго диагонального элемента {Det}')
print(f'Количество верных значащих цифр: {significant_digit(Det, Det_real)}')
print(f'Относительная погрешность: {abs(Det_real - Det) / abs(Det_real)}\n')

# Меняем все диагональные элементы
A = np.array([[9., 5., 6.],[17., 9., 11.], [7., 4., 5.]])
A[0][0] *= 1.01
A[1][1] *= 1.01
A[2][2] *= 1.01
Det = determinat(A)
print(f'Определитель матрицы с погрешностью у всех диагональных элементов {Det}')
print(f'Количество верных значащих цифр: {significant_digit(Det, Det_real)}')
print(f'Относительная погрешность: {abs(Det_real - Det) / abs(Det_real)}')
```

```python
from IPython.display import display, Latex

# функция по нахождению машинного нуля, бесконечности и эпсилон
def machine_number(data_type):
    if data_type == np.single:
        dt = 'одинарной'
    elif data_type == np.double:
        dt = 'двойной'
    else:
        dt = 'расширенной'
    zero = data_type(1)
    infinity = data_type(1)
    epsilon = data_type(1)
    
    # Машинный ноль
    exponent = 0
    while zero != 0:
        zero = data_type(zero / 2)
        exponent -= 1
    display(Latex('\nМашинный ноль в режиме '+ dt +' точности равен $2^{%s}$' % exponent))

    # Машинная бесконечность
    exponent = 0
    while infinity != np.inf:
        infinity = data_type(infinity * 2)
        exponent += 1
    display(Latex('Машинная бесконечность в режиме '+ dt +' точности равен $2^{%s}$' % exponent))

    # Машинный эпсилон
    exponent = 0
    while data_type(1 + epsilon) != 1:
        epsilon = data_type(epsilon / 2)
        exponent -= 1
    display(Latex('Машинный эпсилон в режиме '+ dt +' точности равен $2^{%s}$' % exponent))
```

```python
machine_number(np.single)
machine_number(np.double)
machine_number(np.longdouble)
```

```python
# Функция для ограничения разрядности
def bit_depth_restriction(number: int, m: int):
    tmp = round(number * 10**m)
    return tmp / 10**m
```

```python
# Решение задачи 1 для случая ограничения разрядности
# Вычисляем частичную сумму ряда
def series_sum(n: int, m: int):
    res = 0
    for i in range(n+1):
        res += bit_depth_restriction(36/(i**2 + 7*i + 10), m)
        res = bit_depth_restriction(res, m)
    return res


# Кладем частичные суммы, погрешности в массив
part_sum = []
abs_error = []
for m in range(4,9):
    part_sum.append(series_sum(10000, m))
    abs_error.append(abs(13 - part_sum[-1]))

# Выводим ответы
for i in range(5):
    print(f'Частичная сумма, m = {i + 4} : {part_sum[i]}')
print('\n')
for i in range(5):
    print(f'Погрешность, m = {i + 4} : {abs_error[i]}')
plt.plot(np.arange(4,9), abs_error)
x_ticks = np.arange(4,9)
x_labels = np.arange(4,9)
plt.xticks(ticks=x_ticks, labels=x_labels)
plt.xlabel('Количество знаков после запятой')
plt.ylabel('Величина абсолютной погрешности')
plt.title('Зависимость погрешности суммы ряда от количества разрядов')
plt.savefig('/Users/kirill/edu/3course/Numerical_methods/laba_1/restriction_bit.png')
None
```
